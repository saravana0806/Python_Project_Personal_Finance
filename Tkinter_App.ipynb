```python
'''Entry Widget: entry = tk.Entry(root) creates a text entry widget. entry.pack() adds it to the window.

Button Widget: button = tk.Button(root, text="Click Me", command=on_button_click) creates a button. 
The command parameter specifies the function to call when the button is clicked.

Button Click Event: def on_button_click(): defines the function to execute when the button is clicked. 
It retrieves the text from the entry widget and updates the label with a personalized message.
'''
```

```python
# database.py

import sqlite3

def init_db():
    conn = sqlite3.connect('tasks.db')
    c = conn.cursor()
    c.execute('''
        CREATE TABLE IF NOT EXISTS tasks (
            id INTEGER PRIMARY KEY,
            title TEXT NOT NULL,
            description TEXT,
            due_date TEXT
        )
    ''')
    conn.commit()
    conn.close()

def add_task(title, description, due_date):
    conn = sqlite3.connect('tasks.db')
    c = conn.cursor()
    c.execute('INSERT INTO tasks (title, description, due_date) VALUES (?, ?, ?)', (title, description, due_date))
    conn.commit()
    conn.close()

def get_tasks():
    conn = sqlite3.connect('tasks.db')
    c = conn.cursor()
    c.execute('SELECT * FROM tasks')
    tasks = c.fetchall()
    conn.close()
    return tasks

def update_task(task_id, title, description, due_date):
    conn = sqlite3.connect('tasks.db')
    c = conn.cursor()
    c.execute('UPDATE tasks SET title = ?, description = ?, due_date = ? WHERE id = ?', (title, description, due_date, task_id))
    conn.commit()
    conn.close()

def delete_task(task_id):
    conn = sqlite3.connect('tasks.db')
    c = conn.cursor()
    c.execute('DELETE FROM tasks WHERE id = ?', (task_id,))
    conn.commit()
    conn.close()
```

```python
# data_processing.py

from database import add_task, get_tasks, update_task, delete_task

def process_add_task(title, description, due_date):
    add_task(title, description, due_date)

def process_get_tasks():
    return get_tasks()

def process_update_task(task_id, title, description, due_date):
    update_task(task_id, title, description, due_date)

def process_delete_task(task_id):
    delete_task(task_id)
```

```python
# ui.py

import tkinter as tk
from tkinter import ttk, messagebox
from data_processing import process_add_task, process_get_tasks, process_update_task, process_delete_task
from database import init_db

class TaskManagerApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Task Manager")
        self.root.geometry("800x600")

        self.create_widgets()

    def create_widgets(self):
        notebook = ttk.Notebook(self.root)
        notebook.pack(expand=True, fill='both')

        self.create_add_task_tab(notebook)
        self.create_view_tasks_tab(notebook)

    def create_add_task_tab(self, notebook):
        add_tab = ttk.Frame(notebook)
        notebook.add(add_tab, text="Add Task")

        ttk.Label(add_tab, text="Title:").grid(row=0, column=0, padx=10, pady=10)
        self.title_entry = ttk.Entry(add_tab)
        self.title_entry.grid(row=0, column=1, padx=10, pady=10)

        ttk.Label(add_tab, text="Description:").grid(row=1, column=0, padx=10, pady=10)
        self.description_entry = ttk.Entry(add_tab)
        self.description_entry.grid(row=1, column=1, padx=10, pady=10)

        ttk.Label(add_tab, text="Due Date:").grid(row=2, column=0, padx=10, pady=10)
        self.due_date_entry = ttk.Entry(add_tab)
        self.due_date_entry.grid(row=2, column=1, padx=10, pady=10)

        add_button = ttk.Button(add_tab, text="Add Task", command=self.add_task)
        add_button.grid(row=3, column=1, padx=10, pady=10)

    def create_view_tasks_tab(self, notebook):
        view_tab = ttk.Frame(notebook)
        notebook.add(view_tab, text="View Tasks")

        self.tasks_tree = ttk.Treeview(view_tab, columns=("ID", "Title", "Description", "Due Date"), show="headings")
        self.tasks_tree.heading("ID", text="ID")
        self.tasks_tree.heading("Title", text="Title")
        self.tasks_tree.heading("Description", text="Description")
        self.tasks_tree.heading("Due Date", text="Due Date")
        self.tasks_tree.pack(expand=True, fill='both')

        self.refresh_tasks()

        edit_button = ttk.Button(view_tab, text="Edit Task", command=self.edit_task)
        edit_button.pack(side=tk.LEFT, padx=10, pady=10)

        delete_button = ttk.Button(view_tab, text="Delete Task", command=self.delete_task)
        delete_button.pack(side=tk.RIGHT, padx=10, pady=10)

    def add_task(self):
        title = self.title_entry.get()
        description = self.description_entry.get()
        due_date = self.due_date_entry.get()
        process_add_task(title, description, due_date)
        messagebox.showinfo("Info", "Task added successfully!")
        self.clear_entries()
        self.refresh_tasks()

    def edit_task(self):
        selected_item = self.tasks_tree.selection()[0]
        task_id = self.tasks_tree.item(selected_item, "values")[0]
        self.edit_task_window(task_id)

    def edit_task_window(self, task_id):
        task = None
        for item in process_get_tasks():
            if item[0] == int(task_id):
                task = item
                break

        if task:
            edit_window = tk.Toplevel(self.root)
            edit_window.title("Edit Task")
            edit_window.geometry("400x300")

            ttk.Label(edit_window, text="Title:").grid(row=0, column=0, padx=10, pady=10)
            title_entry = ttk.Entry(edit_window)
            title_entry.grid(row=0, column=1, padx=10, pady=10)
            title_entry.insert(0, task[1])

            ttk.Label(edit_window, text="Description:").grid(row=1, column=0, padx=10, pady=10)
            description_entry = ttk.Entry(edit_window)
            description_entry.grid(row=1, column=1, padx=10, pady=10)
            description_entry.insert(0, task[2])

            ttk.Label(edit_window, text="Due Date:").grid(row=2, column=0, padx=10, pady=10)
            due_date_entry = ttk.Entry(edit_window)
            due_date_entry.grid(row=2, column=1, padx=10, pady=10)
            due_date_entry.insert(0, task[3])

            def save_task():
                new_title = title_entry.get()
                new_description = description_entry.get()
                new_due_date = due_date_entry.get()
                process_update_task(task_id, new_title, new_description, new_due_date)
                messagebox.showinfo("Info", "Task updated successfully!")
                edit_window.destroy()
                self.refresh_tasks()

            save_button = ttk.Button(edit_window, text="Save", command=save_task)
            save_button.grid(row=3, column=1, padx=10, pady=10)

    def delete_task(self):
        selected_item = self.tasks_tree.selection()[0]
        task_id = self.tasks_tree.item(selected_item, "values")[0]
        process_delete_task(task_id)
        messagebox.showinfo("Info", "Task deleted successfully!")
        self.refresh_tasks()

    def refresh_tasks(self):
        for item in self.tasks_tree.get_children():
            self.tasks_tree.delete(item)

        for task in process_get_tasks():
            self.tasks_tree.insert("", tk.END, values=task)

    def clear_entries(self):
        self.title_entry.delete(0, tk.END)
        self.description_entry.delete(0, tk.END)
        self.due_date_entry.delete(0, tk.END)

def run_app():
    init_db()
    root = tk.Tk()
    app = TaskManagerApp(root)
    root.mainloop()
```
